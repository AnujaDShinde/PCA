# PCA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
wine = pd.read_csv('wine.csv')
wine
```

```python
wine['Type'].value_counts()
```

```python
wine1 = wine.iloc[:,1:]
wine1
```

```python
wine1.info()
```

```python
# Converting data to numpy array

wine_array = wine1.values
wine_array
```

```python
# Normalizing the numerical data 

wine_norm = scale(wine_array)
wine_norm
```

```python
# Continue with the transformation of the data onto unit scale (mean=0 and variance=1), 
# which is a requirement for the optimal performance of many machine learning algorithms.
```

```python
# Applying PCA

pca = PCA(n_components = 13)
wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Component Matrix or Covariance Matrix

pca.components_
```

```python
# Variance for each PCA

var = pca.explained_variance_ratio_
var
```

```python
# The first principal component contains 36.19 % of the information.
# The second principal component contains 19.20 % of the variance and the third principal component contains 11.12 % of the variance.
# The remaining principal component contain the rest of the variance of the dataset as mentioned above.
```

```python
# Cumulative variance for each PCA
var1 = np.cumsum(np.round(var,4)*100)
var1
```

```python
# The first principal component contains 36.2 % of the information
# First two principal component contain 55.41 % of the information and so on.....
```

```python
# Variance plot for obtained PCA components  

plt.plot(var1, color = 'orange')
```

```python
final_df = pd.concat([wine['Type'], pd.DataFrame(wine_pca[:,0:3], columns = ['PC1','PC2','PC3'])], axis = 1)
final_df
```

```python
# Visualization of PCA using scatter plot

fig = plt.figure(figsize = (18,12))
sns.scatterplot(data = final_df)
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
```

```python
# K-means algorithm aims to choose centroids that minimise the inertia
# Random state can be anything from 0 to 42, but the same number to be used everytime, so that the results won't change. 

wcss = []
for i in range (1,6):
    kmeans = KMeans(n_clusters = i, random_state = 2)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K 

plt.plot(range(1,6), wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Choose k = 3

clusters3 = KMeans(3, random_state = 45).fit(wine_norm)
clusters3
```

```python
clusters3.labels_
```

```python
wine2 = wine.copy()
wine2['Clusterid'] = clusters3.labels_
wine2
```

```python
wine2['Clusterid'].value_counts()
```

# Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# Create dendrogram

plt.figure(figsize = (18,8))
dendrogram = sch.dendrogram(sch.linkage(wine_norm, 'complete'))
```

```python
h_clusters = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
h_clusters
```

```python
y = pd.DataFrame(h_clusters.fit_predict(wine_norm), columns = ['Clusterid'])
y['Clusterid'].value_counts()
```

```python
wine3 = wine.copy()
wine3['Clusterid'] = h_clusters.labels_
wine3
```

```python
# From above output we can conclude that in KMeans Clustering and in Hierarchical Clustering,  
# 62 and 58 records belong to cluster 0. 51 and 56 records belong to cluster 1. 65 and 64 records belong to cluster 2
# We have obtained same and optimum number of clusters with the original data.
```
